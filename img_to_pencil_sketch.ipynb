```python
#Importing Necessary Libraries
import cv2
import matplotlib.pyplot as plt
```

```python
#Reading an Image
image=cv2.imread("Downloads/tom_and_jerry.jpg")

#Converting BGR to RGB
RGB_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

#Converting original image to grey scale
grey_image=cv2.cvtColor(RGB_image,cv2.COLOR_BGR2GRAY)

#Inverting grey scale image
invert_image=cv2.bitwise_not(grey_image)

#Blurring Inverted Image by using GaussianBlur
blurred_image=cv2.GaussianBlur(invert_image,(25,25),0)

#Again Inverting the blurred Image
inverted_blur_image=cv2.bitwise_not(blurred_image)

#Final pencil image
Pencil_sketch_image=cv2.divide(grey_image,inverted_blur_image,scale=250.0)
```

```python
plt.figure(figsize=(14,8))

plt.subplot(1,2,1)
plt.title("Original Image",size=18)
plt.imshow(RGB_image)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Sketch_image')
rgb_sketch=cv2.cvtColor(Pencil_sketch_image,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
